```python
import os

# Define the path to the directory containing images
folder_path = 'imgs'

if os.path.exists(folder_path):
    # List all items in the directory and filter to keep only files
    files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    
    # Sort files alphabetically to ensure deterministic order (e.g. sclimg1, sclimg2, ...)
    files.sort()
    
    print(f"Found {len(files)} files to rename in '{folder_path}'.\n")
    
    # Iterate and rename each file
    for idx, filename in enumerate(files, start=1):
        # Get the file extension (e.g., .jpeg, .png)
        _, ext = os.path.splitext(filename)
        
        # Define the new filename format
        new_name = f"sclimg{idx}{ext}"
        
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, new_name)
        
        # Perform the rename operation
        os.rename(old_path, new_path)
        print(f"Renamed: '{filename}' -> '{new_name}'")
        
    print("\nAll files have been successfully renamed.")
else:
    print(f"Error: The folder '{folder_path}' does not exist.")
```
